# *Download output*

```python
import os
import shutil
from IPython.display import FileLink

# Définir le répertoire de travail
os.chdir('/kaggle/working')

# Définir le dossier à compresser et le nom du fichier ZIP
folder_to_zip = 'BOA-dataset-treatment'
output_zip = '/kaggle/working/BOA-dataset-treatment.zip'

# Vérifier que le dossier existe
if os.path.exists(folder_to_zip):
    # Créer l'archive ZIP
    shutil.make_archive('/kaggle/working/BOA-dataset-treatment', 'zip', folder_to_zip)
    
    # Vérifier que le fichier ZIP a été créé
    if os.path.exists(output_zip):
        # Générer un lien de téléchargement
        display(FileLink('BOA-dataset-treatment.zip'))
    else:
        print("Erreur : Le fichier ZIP n'a pas été créé.")
else:
    print(f"Erreur : Le dossier {folder_to_zip} n'existe pas.")
```
